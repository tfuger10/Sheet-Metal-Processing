First we will take the folder with all of the images and separate them into folders with each class

In order for the script below to work:

- The csv file for the script below has to be formatted with the first column indicating the image name and the second column indicating the class
- The new directory folder needs to be created prior to running this script
- The full directories need to be filled out in the cell below

```python
from __future__ import print_function
import pandas as pd
import shutil
import os
import sys
import trimesh
```

```python
original_dir = r"C:\Users\Tim\Documents\Flatiron\Capstone\Part-Processing\Meshes\0_FabInc_NoClass"
new_dir = r"C:\Users\Tim\Documents\Flatiron\Capstone\Part-Processing\Meshes\1A_FabInc_OBJ"
```

```python
# This script was taken from https://github.com/bhanuprathap2000/dsp/blob/master/datasplit.py

labels = pd.read_csv(r'C:\Users\Tim\Documents\Flatiron\Capstone\Part-Processing\Meshes\Data_Mesh_Target.csv')

if not os.path.exists(new_dir):
    os.mkdir(new_dir)

for filename, class_name in labels.values:
    # Create subdirectory with `class_name`
    if not os.path.exists(new_dir + '/' + str(class_name)):
        os.mkdir(new_dir + '/' + str(class_name))
    src_path = original_dir + '/'+ filename + '.stl'
    dst_path = new_dir + '/' + str(class_name) + '/' + filename + '.obj'
    try:
        file = trimesh.load_mesh(src_path)
        file_obj = trimesh.exchange.obj.export_obj(file, header=None)
        with open(dst_path, 'w') as f:
            f.write(file_obj)
        print(filename + ": sucessful")
    except IOError as e:
        print('Unable to export file {} to {}'
              .format(src_path, dst_path))
    except:
        print('When try to export file {} to {}, unexpected error: {}'
              .format(src_path, dst_path, sys.exc_info()))
```

Now we must split the meshes into train, test, and validation folders.

```python
import os
import splitfolders

def split_data(input_dir, output_dir, x):
    splitfolders.ratio(input_dir, output = output_dir, seed = 89, ratio = (x, (1-x)/2, (1-x)/2), group_prefix = None)

split_data(r'C:\Users\Tim\Documents\Flatiron\Capstone\Part-Processing\Meshes\1A_FabInc_OBJ_Class', r'C:\Users\Tim\Documents\Flatiron\Capstone\Part-Processing\Meshes\1B_FabInc', 0.8)
```
